```python
#lire les donn√©e 
data = pd.read_csv('train.csv')
data.head()
data.info()

#make a PCA to reduce the dimensionality of the data with these column :  
# 0   PassengerId   8693 non-null   object 
#  1   HomePlanet    8492 non-null   object 
#  2   CryoSleep     8476 non-null   object 
#  3   Cabin         8494 non-null   object 
#  4   Destination   8511 non-null   object 
#  5   Age           8514 non-null   float64
#  6   VIP           8490 non-null   object 
#  7   RoomService   8512 non-null   float64
#  8   FoodCourt     8510 non-null   float64
#  9   ShoppingMall  8485 non-null   float64
#  10  Spa           8510 non-null   float64
#  11  VRDeck        8505 non-null   float64
#  12  Name          8493 non-null   object 
#  13  Transported   8693 non-null   bool  

from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
from sklearn.metrics import classification_report
from sklearn.metrics import confusion_matrix
from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import OneHotEncoder
from sklearn.preprocessing import OneHotEncoder
import matplotlib.pyplot as plt


# Create an instance of the OneHotEncoder
encoder = OneHotEncoder()

# Create an instance of the PCA class
pca = PCA()

# Create an instance of the StandardScaler
scaler = StandardScaler()

# Encode categorical variables
data_encoded = encoder.fit_transform(data[['HomePlanet', 'CryoSleep', 'Cabin', 'Destination', 'VIP', 'Name']])
data_encoded = pd.DataFrame(data_encoded.toarray())
print(data_encoded)



```

```python
# Scale the data
data_scaled = scaler.fit_transform(data_encoded)

# Fit the PCA
pca.fit(data_scaled)

# Transform the data
data_pca = pca.transform(data_scaled)
print(data_pca)
```

```python
# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(data_pca, data['Transported'], test_size=0.2, random_state=42)

# Create an instance of the RandomForestClassifier
rf = RandomForestClassifier()

# Fit the model
rf.fit(X_train, y_train)

# Make predictions
y_pred = rf.predict(X_test)

# Calculate the accuracy
accuracy = accuracy_score(y_test, y_pred)
print('Accuracy:', accuracy)

# Print the classification report
print(classification_report(y_test, y_pred))

# Print the confusion matrix
print(confusion_matrix(y_test, y_pred))

#plot the explained variance ratio
plt.plot(pca.explained_variance_ratio_)
plt.xlabel('Number of components')
plt.ylabel('Explained variance ratio')
plt.show()

#plot the cumulative explained variance ratio
plt.plot(np.cumsum(pca.explained_variance_ratio_))
plt.xlabel('Number of components')
plt.ylabel('Cumulative explained variance ratio')
plt.show()
```
